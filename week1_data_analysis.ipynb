```python
years = list(map(int, range(1983,2014)))
years
```

```python
India.plot()
plt.title('Indian Emmigrants to Canada')
plt.xlabel('Years')
plt.ylabel('No. of emmigrants')
plt.show()
```

```python
China.plot()
plt.title('Chinese Emmigrants to Canada')
plt.xlabel('Years')
plt.ylabel('No. of emmigrants')
plt.show()
```

```python
df_IndChina = df_can.loc[['India','China'], years]
df_IndChina.head()
```
